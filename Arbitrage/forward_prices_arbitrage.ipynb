# Variables:

T: time until delivery date in a forward or futures contract (in years)\
S0: price of the asset underlying the forward or futures contract today \
F0: forward or futures price today \
r: zero coupon rish free rate of interest per annum, continuous compounding  for investment maturing at the delivery date \
I: value at te present time of the income of an investment \
q: yield at time T

```python
def forward_prices_arbitrage(F0,S0,r,T,short_selling: bool): 
    if short_selling == True:
        if F0> S0*np.exp(r*T):
            print('borrow at r rate for T years,buy the asset and short the forward on the asset, profit:', F0 - S0*np.exp(r*T))
        elif F0< S0*np.exp(r*T):
            print('short the asset, invest the money and long the forward contract on it, profit:', S0*np.exp(r*T)-F0)
        else:
            print('No possible arbitrage in this situation')
    else:
        if F0> S0*np.exp(r*T):
            print('borrow S0 at r rate for T years,enter a forward to sell. AT time T, sell F0, repay loan for S0exp(rT), profit:', F0 - S0*np.exp(r*T))
        elif F0< S0*np.exp(r*T):
            ## If short-selling is not possible, you should own the stock to gain profit from selling it
            print('IF YOU OWN the asset: Sell it at S0, invest the proceeds at ratr r for T years, enter a forward to buy the asset,profit:', S0*np.exp(r*T)-F0)
        else:
            print('No possible arbitrage in this situation')
```

```python
def forward_prices_with_known_income_arbitrage(F0,S0,r,T,I,short_selling: bool): 
    if short_selling == True:
        if F0> (S0-I)*np.exp(r*T):
            print('borrow I at r rate for T_i years and S0-I at rate r for T years,buy the asset and short the forward on the asset. At time T_i, use the payment of I to repay the first loan. AT time T, sell the asset, repay the second loan, profit:', F0 - (S0-I)*np.exp(r*T))
        elif F0< (S0-I)*np.exp(r*T):
            print('short the asset, invest I at r for T_i years and the rest at r for T years and long the forward contract on it. At itme T_i pay the income from the first investment. At time T, buy the asset, profit:', (S0-I)*np.exp(r*T)-F0)
        else:
            print('No possible arbitrage in this situation')
    else:
        if F0> (S0-I)*np.exp(r*T):
            print('borrow S0 at r rate for T years,enter a forward to sell. AT time T, sell F0, repay loan for S0exp(rT), profit:', F0 - (S0-I)*np.exp(r*T))
        elif F0< (S0-I)*np.exp(r*T):
            ## If short-selling is not possible, you should own the stock to gain profit from selling it
            print('IF YOU OWN the asset: Sell it at S0, invest the proceeds at rate r for T years, enter a forward to buy the asset,profit:', (S0-I)*np.exp(r*T)-F0)
        else:
            print('No possible arbitrage in this situation')
```
