```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import yfinance as yf
```

```python
df=pd.read_csv('sp500_prices.csv')
df.head()
```

```python
def portfolio_returns(weights, returns):
    return np.dot(weights, returns)
```

```python
AMD_returns = df['AMD']
AZO_returns = df['AZO']
ptf=[0.6, 0.4]
ptf_returns= portfolio_returns(ptf, [AMD_returns, AZO_returns])
plt.plot(ptf_returns,label='ptf')
plt.plot(AMD_returns, label='AMD')
plt.plot(AZO_returns, label='AZO')
plt.legend()
plt.show()
```

```python
def calculate_beta(stock_returns, market_returns):
    covariance = np.cov(stock_returns, market_returns)[0][1]
    market_variance = np.var(market_returns)
    beta = covariance / market_variance
    return beta
```

```python
beta_AMD = calculate_beta(AMD_returns, ptf_returns)
beta_AZO = calculate_beta(AZO_returns, ptf_returns)
print(f"Beta of AMD: {beta_AMD}")
print(f"Beta of AZO: {beta_AZO}")
```
