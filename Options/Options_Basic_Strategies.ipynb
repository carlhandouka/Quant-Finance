```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def bull_spread_payoff(S, K1, K2): #long call K1, short call K2
    return np.maximum(0, S - K1) - np.maximum(0, S - K2)
```

```python
def bear_spread_payoff(S, K1, K2): #short put K1, long put K2
    return np.maximum(0, K2 - S) - np.maximum(0, K1 - S)
```

```python
def box_spread_payoff(S, K1, K2): #long bull spread K1,K2 and long bear spread K1,K2
    return bull_spread_payoff(S, K1, K2) + bear_spread_payoff(S, K1, K2)
```

```python
def butterfly_spread_payoff(S, K1, K2, K3): #long call K1, short 2 calls K2, long call K3
    return (np.maximum(0, S - K1) - 2 * np.maximum(0, S - K2) + np.maximum(0, S - K3))
```

```python
def straddle_payoff(S, K): #long call and long put at strike K
    return np.maximum(0, S - K) + np.maximum(0, K - S)
```

```python
def strangle_payoff(S, K1, K2): #long put K1, long call K2
    return np.maximum(0, S - K2) + np.maximum(0, K1 - S)
```

```python
def strips_payoff(S, K): #long 2 puts and short 1 call at strike K
    return 2 * np.maximum(0, K - S) - np.maximum(0, S - K)
```

```python
def straps_payoff(S, K): #long 2 calls and short 1 put at strike K
    return 2 * np.maximum(0, S - K) - np.maximum(0, K - S)
```
